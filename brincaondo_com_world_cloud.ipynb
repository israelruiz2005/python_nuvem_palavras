```
# Apenas para brincar
# importando os pacotes
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
from PIL import Image
```

```
#Aqui você usa um data frame eu usei fixo só para brincar
all_summary =" Tudo o que escrevemos, esta escrito, não há borracha que apague, aquilo que você publicou na rede, afinal é uma rede, o seu peixe fica preso, em fios, finos, quase como de ouro fino, mas apague isso, ou aquilo, é ou aquilo, e isso também, tudo o que você deixar no gmail, no hotmail, códificado em UTF, em endereços _http"
```

```
# lista de stopword, ou palavras que não deverão ser utilizadas
stopwords = set(STOPWORDS)
stopwords.update(["da", "meu", "em", "você", "de", "ao", "os", "ai", "UTF", "_http","hotmail", "gmail"])
```

```
# criando a wordcloud

# carregar a imagem, editei uma imagem para ficar melhor
image_mask = np.array(Image.open("/content/brain-1212a.png")) 

# gerar uma wordcloud, basicamente informei qual era a imagem, as stopwords, a cor de fundo, o tamanho comprimento e altura
wordcloud = WordCloud(mask=image_mask,stopwords=stopwords,
                      background_color="black",
                      width=2000, height=1000).generate(all_summary)
 
# mostrar a imagem final
fig, ax = plt.subplots(figsize=(10,6))
ax.imshow(wordcloud, interpolation='bilinear')
ax.set_axis_off()
 
plt.imshow(wordcloud);
```
